```python
## first bullet point workshop 4
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

file_path = 'ThuptenWangpo_Workshop3.csv'
df = pd.read_csv(file_path)

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(df['processed_text'].fillna(''))

tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

tfidf_df
```

```python

## Workshop 4 bullet point 2

ngram_shapes = {}

for i in range(1, 6):
    vectorizer = TfidfVectorizer(ngram_range=(1, i))
    
    tfidf_matrix = vectorizer.fit_transform(df['processed_text'].fillna(''))
    
    ngram_shapes[i] = tfidf_matrix.shape
    print(f'ngram_range=(1, {i}): Shape = {tfidf_matrix.shape}')

print("\nOutput shapes for different ngram shapes:")
for n, shape in ngram_shapes.items():
    print(f"ngram_range=(1, {n}): {shape}")
```
